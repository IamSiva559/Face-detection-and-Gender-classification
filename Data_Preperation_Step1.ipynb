# Detecting face using Haar CascadeClassifier
https://github.com/opencv/opencv/blob/master/data/haarcascades/haarcascade_frontalface_default.xml

1. Detecting face using Haar CascadeClassifier
2. croping faces 
3. saving them to local 

```python
def detect_face_and_do_crop(img_path,gender,i):
    img = cv2.imread(img_path)
    haar = cv2.CascadeClassifier('./data/haarcascade_frontalface_default.xml')
    face = haar.detectMultiScale(img,1.3,5)
    for x,y,w,h in face:
        cropped_img = img[y:y+h,x:x+w]
        if gender == 'male':
            cv2.imwrite('./data/crop/male_crop/{}_{}.png'.format(gender,i),cropped_img)
        else:
            cv2.imwrite('./data/crop/female_crop/{}_{}.png'.format(gender,i),cropped_img)
```

```python
for i,path in enumerate(femalepath):
    try:
        detect_face_and_do_crop(path,'female',i)
        print('INFO: {}/{} processed sucessfully'.format(i,len(femalepath)))
        
    except:
        print('INFO: {}/{} cannot be processed'.format(i,len(femalepath)))   
        
```

```python
for i,path in enumerate(malepath):
    try:

        detect_face_and_do_crop(path,'male',i)
        print('INFO: {}/{} processed sucessfully'.format(i,len(malepath)))
        
    except:
        print('INFO: {}/{} cannot be processed'.format(i,len(malepath)))
```

# So, from analysis i will resize all the image in 100 x 100 form and remove image with size less or equal 54

```python
crop_imgs_df_new = crop_imgs_df[crop_imgs_df['size'] > 60]
```

```python
crop_imgs_df_new.head(5)
```

```python
#Extracting Gender from image path
string = crop_imgs_df_new['img_path'][0]
def gender(string):
    try:

        return string.split('_')[0].split('/')[-1]
    except:
        return None
```

```python
crop_imgs_df_new['gender'] = crop_imgs_df_new['img_path'].apply(gender)
```

conclusion : resizing all the image in df_new to 100 x 100 array

```python
def resize_img(path_to_resize):
    try:

        # step - 1: read image
        img = cv2.imread(path_to_resize)
        # step - 2: convert into grayscale
        gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
        # step -3: resize into 100 x 100 array
        size = gray.shape[0]
        if size >= 100: #shrink
            gray_re = cv2.resize(gray,(100,100),cv2.INTER_AREA) # SHRINK
        else: # enlarge
            gray_re = cv2.resize(gray,(100,100),cv2.INTER_CUBIC) # ENLARGE
        # step -4: Flatten Image (1x10,000)
        return gray_re.flatten()
    except:
        return None
```

```python
# structuring function
crop_imgs_df_new['structure_data'] = crop_imgs_df_new['img_path'].apply(resize_img)
```

```python
# copy and expand their columns
structure_df = crop_imgs_df_new['structure_data'].apply(pd.Series)
```

```python
structure_df =  pd.concat((crop_imgs_df_new['gender'],structure_df),axis=1)
```
